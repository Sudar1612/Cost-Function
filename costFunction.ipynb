## Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse
```

```python
experience = [1.2,1.5,1.9,2.2,2.4,2.5,2.8,3.1,3.3,3.7,4.2,4.4] 
salary = [1.7,2.4,2.3,3.1,3.7,4.2,4.4,6.1,5.4,5.7,6.4,6.2] 
```

```python
df=pd.DataFrame({"Salary":salary,"Experience":experience})
df.head()
```

## Starting the line

```python
plt.scatter(df.Experience,df.Salary,label='Data Points')
plt.xlim(1,5)
plt.ylim(1,7)
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
```

Equation for linear regression : y=mx+b

```python
# let us randomly assign a value for m and b 
# starting with small values for m and b

m=0.1
b=1.1

line1=[] #to store prediced predicted values with m=0.1 and b=1.1

#here salary is the output value

#geneate prediction for every data
for i in range(len(df)):
    sal=m*df.Experience[i]+b
    line1.append(sal)
```

```python
#plotting predicted values

plt.scatter(df.Experience,df.Salary,label='Actual')
plt.plot(df.Experience,line1,color='red',label='Predicted')
plt.xlim(1,5)
plt.ylim(1,7)
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
```

```python
#calculating the mse for the predicted line
mse_line1=mse(df.Salary,line1)
mse_line1
```

This mse value is so high and the slope for the predicted line is so low , hence let's try increasing m having b as constant

## Cost Fcuntion over a range of m (Theta)

```python
def Error_func(theta,df):
    b=1.1 # b is constant

    predicted_salary=[] 

    #geneate prediction for every data
    for i in range(len(df)):
        predicted_salary.append(theta*df.Experience[i]+b)
    
    MSE=mse(df.Salary,predicted_salary)
    
    return MSE
```

```python
#trying all the values of m (theta) to find the optimal value i.e the one that has the lowest mse

#range of slopes from 0 to 1.5 with an increment of 0.01

slope_range=[i/100 for i in range(0,150)]
Cost=[]
for slope in slope_range:
    cost=Error_func(slope,df)
    Cost.append(cost)
```

## Visualizing Cost w.r.t Theta

```python
# making cost and theta as dataframe

Cost_df=pd.DataFrame({
    'Theta' : slope_range,
    'Cost' : Cost
})
```

```python
plt.plot(Cost_df.Theta,Cost_df.Cost,label='Cost Function Curve')
plt.xlabel('Theta')
plt.ylabel('Cost Fucntion')
plt.legend()
```
